```python
# Answer the following guiding questions in your notebook:

# LOSS RATIO ANALYSIS


# ------------------------------------------
# DATA CLEANING
# ------------------------------------------

# Keep required columns
loss_df = corrected_df[
    [
        'Province',
        'VehicleType',
        'Gender',
        'TotalPremium',
        'TotalClaims'
    ]
].copy()

# ------------------------------------------
# OVERALL LOSS RATIO
# ------------------------------------------

total_claims = loss_df['TotalClaims'].sum()
total_premium = loss_df['TotalPremium'].sum()

overall_loss_ratio = (
    total_claims / total_premium
) * 100

print(f"\nOverall Loss Ratio: {overall_loss_ratio:.2f}%")

# ------------------------------------------
# LOSS RATIO BY PROVINCE
# ------------------------------------------

province_loss = (
    loss_df
    .groupby('Province')
    .agg({
        'TotalClaims': 'sum',
        'TotalPremium': 'sum'
    })
)

province_loss['LossRatio'] = (
    province_loss['TotalClaims'] /
    province_loss['TotalPremium']
) * 100

print("\nLoss Ratio by Province")
print(province_loss[['LossRatio']])

# Plot
plt.figure(figsize=(10,6))

sns.barplot(
    x=province_loss.index,
    y=province_loss['LossRatio']
)

plt.xticks(rotation=45)

plt.title('Loss Ratio by Province')
plt.ylabel('Loss Ratio (%)')

plt.show()

# ------------------------------------------
# LOSS RATIO BY VEHICLE TYPE
# ------------------------------------------

vehicle_loss = (
    loss_df
    .groupby('VehicleType')
    .agg({
        'TotalClaims': 'sum',
        'TotalPremium': 'sum'
    })
)

vehicle_loss['LossRatio'] = (
    vehicle_loss['TotalClaims'] /
    vehicle_loss['TotalPremium']
) * 100

print("\nLoss Ratio by VehicleType")
print(vehicle_loss[['LossRatio']])

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=vehicle_loss.index,
    y=vehicle_loss['LossRatio']
)

plt.xticks(rotation=45)

plt.title('Loss Ratio by Vehicle Type')
plt.ylabel('Loss Ratio (%)')

plt.show()

# ------------------------------------------
# LOSS RATIO BY GENDER
# ------------------------------------------

gender_loss = (
    loss_df
    .groupby('Gender')
    .agg({
        'TotalClaims': 'sum',
        'TotalPremium': 'sum'
    })
)

gender_loss['LossRatio'] = (
    gender_loss['TotalClaims'] /
    gender_loss['TotalPremium']
) * 100

print("\nLoss Ratio by Gender")
print(gender_loss[['LossRatio']])

# Plot
plt.figure(figsize=(6,5))

sns.barplot(
    x=gender_loss.index,
    y=gender_loss['LossRatio']
)

plt.title('Loss Ratio by Gender')
plt.ylabel('Loss Ratio (%)')

plt.show()
```

```python
# FINANCIAL VARIABLE DISTRIBUTION ANALYSIS

# AUTOMATICALLY SELECT NUMERICAL FEATURES

numerical_cols = corrected_df.select_dtypes(
    include=['int64', 'float64']
).columns

# ------------------------------------------
# SELECT KEY FINANCIAL VARIABLES
# ------------------------------------------

financial_keywords = [
    'Premium',
    'Claim',
    'Value',
    'Insured',
    'Outstanding'
]

financial_features = [
    col for col in numerical_cols
    if any(keyword.lower() in col.lower()
           for keyword in financial_keywords)
]

print("Financial Variables:")
print(financial_features)

# ------------------------------------------
# CLEAN DATA
# ------------------------------------------

finance_df = corrected_df[financial_features].copy()

# Remove missing values
finance_df = finance_df.dropna()

# ------------------------------------------
# DISTRIBUTION PLOTS
# ------------------------------------------

for col in financial_features:

    plt.figure(figsize=(8,5))

    sns.histplot(
        finance_df[col],
        kde=True,
        bins=30
    )

    plt.title(f'Distribution of {col}')

    plt.xlabel(col)

    plt.ylabel('Frequency')

    plt.show()

# ------------------------------------------
# BOX PLOTS FOR OUTLIERS
# ------------------------------------------

for col in financial_features:

    plt.figure(figsize=(8,4))

    sns.boxplot(x=finance_df[col])

    plt.title(f'Box Plot of {col}')

    plt.xlabel(col)

    plt.show()

# ------------------------------------------
# IQR-BASED OUTLIER DETECTION
# ------------------------------------------

print("\n========== OUTLIER SUMMARY ==========")

for col in financial_features:

    # Quartiles
    Q1 = finance_df[col].quantile(0.25)
    Q3 = finance_df[col].quantile(0.75)

    # IQR
    IQR = Q3 - Q1

    # Bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = finance_df[
        (finance_df[col] < lower_bound) |
        (finance_df[col] > upper_bound)
    ]

    # Percentage of outliers
    pct_outliers = (
        len(outliers) / len(finance_df)
    ) * 100

    print(f"\nFeature: {col}")
    print(f"Outliers: {len(outliers)}")
    print(f"Percentage: {pct_outliers:.2f}%")
    print(f"Lower Bound: {lower_bound:.2f}")
    print(f"Upper Bound: {upper_bound:.2f}")
```

```python
# TEMPORAL TREND ANALYSIS
# Claim Frequency and Severity
# DATE PREPARATION

# Convert TransactionMonth to datetime
corrected_df['TransactionMonth'] = pd.to_datetime(
    corrected_df['TransactionMonth']
)

# Extract monthly period
corrected_df['YearMonth'] = (
    corrected_df['TransactionMonth']
    .dt.to_period('M')
    .astype(str)
)

# ------------------------------------------
# CLAIM INDICATOR
# ------------------------------------------

# Create indicator for claims
corrected_df['HasClaim'] = corrected_df['TotalClaims'] > 0

# ------------------------------------------
# MONTHLY CLAIM FREQUENCY
# ------------------------------------------

monthly_frequency = (
    corrected_df.groupby('YearMonth')
    .agg(
        TotalPolicies=('PolicyID', 'count'),
        ClaimCount=('HasClaim', 'sum')
    )
)

monthly_frequency['ClaimFrequency'] = (
    monthly_frequency['ClaimCount'] /
    monthly_frequency['TotalPolicies']
)

print("\nMonthly Claim Frequency")
print(monthly_frequency.head())

# ------------------------------------------
# PLOT CLAIM FREQUENCY
# ------------------------------------------

plt.figure(figsize=(12,6))

sns.lineplot(
    x=monthly_frequency.index,
    y=monthly_frequency['ClaimFrequency'],
    marker='o'
)

plt.xticks(rotation=45)

plt.title('Monthly Claim Frequency Trend')

plt.ylabel('Claim Frequency')

plt.xlabel('Month')

plt.show()

# ------------------------------------------
# MONTHLY CLAIM SEVERITY
# ------------------------------------------

# Keep only rows with claims
claims_only = corrected_df[corrected_df['TotalClaims'] > 0]

monthly_severity = (
    claims_only.groupby('YearMonth')
    .agg(
        TotalClaims=('TotalClaims', 'sum'),
        NumberOfClaims=('TotalClaims', 'count')
    )
)

monthly_severity['ClaimSeverity'] = (
    monthly_severity['TotalClaims'] /
    monthly_severity['NumberOfClaims']
)

print("\nMonthly Claim Severity")
print(monthly_severity.head())

# ------------------------------------------
# PLOT CLAIM SEVERITY
# ------------------------------------------

plt.figure(figsize=(12,6))

sns.lineplot(
    x=monthly_severity.index,
    y=monthly_severity['ClaimSeverity'],
    marker='o'
)

plt.xticks(rotation=45)

plt.title('Monthly Claim Severity Trend')

plt.ylabel('Average Claim Amount')

plt.xlabel('Month')

plt.show()

# ------------------------------------------
# OPTIONAL: PREMIUM TREND
# ------------------------------------------

monthly_premium = (
    corrected_df.groupby('YearMonth')['TotalPremium']
    .sum()
)

plt.figure(figsize=(12,6))

sns.lineplot(
    x=monthly_premium.index,
    y=monthly_premium.values,
    marker='o'
)

plt.xticks(rotation=45)

plt.title('Monthly Total Premium Trend')

plt.ylabel('Total Premium')

plt.xlabel('Month')

plt.show()
```

```python
# VEHICLE MAKE/MODEL CLAIM ANALYSIS

# DATA CLEANING

vehicle_df = corrected_df[
    ['make', 'Model', 'TotalClaims']
].copy()


# Keep only positive claims
vehicle_df = vehicle_df[
    vehicle_df['TotalClaims'] > 0
]

# ------------------------------------------
# CLAIM ANALYSIS BY VEHICLE MAKE
# ------------------------------------------

make_claims = (
    vehicle_df
    .groupby('make')
    .agg(
        AverageClaim=('TotalClaims', 'mean'),
        TotalClaimAmount=('TotalClaims', 'sum'),
        ClaimCount=('TotalClaims', 'count')
    )
)

# Optional:
# Keep makes with enough observations
make_claims = make_claims[
    make_claims['ClaimCount'] >= 10
]

# ------------------------------------------
# HIGHEST CLAIM MAKES
# ------------------------------------------

highest_makes = (
    make_claims
    .sort_values(
        by='AverageClaim',
        ascending=False
    )
    .head(10)
)

print("\nHighest Claim Vehicle Makes")
print(highest_makes)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=highest_makes.index,
    y=highest_makes['AverageClaim']
)

plt.xticks(rotation=45)

plt.title('Top 10 Vehicle Makes by Average Claim')

plt.ylabel('Average Claim Amount')

plt.xlabel('Vehicle Make')

plt.show()

# ------------------------------------------
# LOWEST CLAIM MAKES
# ------------------------------------------

lowest_makes = (
    make_claims
    .sort_values(
        by='AverageClaim',
        ascending=True
    )
    .head(10)
)

print("\nLowest Claim Vehicle Makes")
print(lowest_makes)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=lowest_makes.index,
    y=lowest_makes['AverageClaim']
)

plt.xticks(rotation=45)

plt.title('Lowest 10 Vehicle Makes by Average Claim')

plt.ylabel('Average Claim Amount')

plt.xlabel('Vehicle Make')

plt.show()

# ==========================================
# CLAIM ANALYSIS BY VEHICLE MODEL
# ==========================================

model_claims = (
    vehicle_df
    .groupby('Model')
    .agg(
        AverageClaim=('TotalClaims', 'mean'),
        TotalClaimAmount=('TotalClaims', 'sum'),
        ClaimCount=('TotalClaims', 'count')
    )
)

# Filter models with enough observations
model_claims = model_claims[
    model_claims['ClaimCount'] >= 10
]

# ------------------------------------------
# TOP MODELS BY CLAIM AMOUNT
# ------------------------------------------

top_models = (
    model_claims
    .sort_values(
        by='AverageClaim',
        ascending=False
    )
    .head(10)
)

print("\nHighest Claim Vehicle Models")
print(top_models)

# Plot
plt.figure(figsize=(14,6))

sns.barplot(
    x=top_models.index,
    y=top_models['AverageClaim']
)

plt.xticks(rotation=45)

plt.title('Top Vehicle Models by Average Claim')

plt.ylabel('Average Claim Amount')

plt.xlabel('Vehicle Model')

plt.show()
```
